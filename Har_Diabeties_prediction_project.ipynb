** Diabetes prediction using Random forests model **

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/diabetes.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!="Outcome"]
y = df.iloc[:,df.columns=="Outcome"]
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain, xtest , ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

```python
xtrain.head(5)
```

```python
ytrain.head(5)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain, ytrain)
```

```python
predicted_output = model.predict(xtest)
```

```python
print(predicted_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_score = accuracy_score(predicted_output, ytest)
print("Accuracy score for Random forest model : ", acc_score)
```
